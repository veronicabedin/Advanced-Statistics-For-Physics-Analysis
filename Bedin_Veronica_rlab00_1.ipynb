```r
file <- "lakes.txt"

rows <- readLines(file)

a <- strsplit(rows, "\\s+") #separator: any number of spaces 

#save headings and remove from the data
#to save the vector in a list you have to use the double [[]]

headings <-a[[1]]
#a[-index] removes that index
a <- a[-1]

for (i in 1:length(a)){
    tmp=a[[i]][1]                       #save first element
    a[[i]] <- a[[i]][-1]                #remove first row 
    a[[i]][1] <- paste(tmp, a[[i]][1]) #add it to the second element, unique string
}
a
```

```r
#Have a look at this 
df <- data.frame(do.call(rbind,a), stringsAsFactors = FALSE)

names(df) <- headings

#convert all columns to double apart from the first one
df[, -1] <- lapply(df[ , -1], as.numeric)

df
```

```r
#1 
#dataframe=list of vectors so again [[]] to access the vector
max_vol <- which.max(df[["Volume"]])
min_vol <- which.min(df[["Volume"]])
max_are <- which.max(df[["Area"]])
max_are <- which.max(df[["Area"]])

#Show with cat 
```
